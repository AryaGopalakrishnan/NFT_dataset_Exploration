# Cleaning

```python
df.isnull().any() #Detected missing value in nft_token_id,erc_standard,original_amount,usd_amount
```

```python
df.isnull().sum() #number of missing value for respective columns
```

```python
df.usd_amount=df['usd_amount'].fillna(0.0) #Replaced null value for usd_amount with 0.0
#df.isnull().any()
```

```python
#df.nft_token_id=df['nft_token_id'].fillna(0.0) #Replaced null value for nft_token_id with 0.0 (doubtful if it is appropriate to replace id)
#df.isnull().any()
```

```python
#df.erc_standard=df['erc_standard'].fillna("") #doubtful if appropriate to replace the standard
#df.isnull().any()
```

```python
df.original_amount=df['original_amount'].fillna(0) #replaced null with 0 for original_amount
df.isnull().any()
```

```python
df['tx_hash']
```

```python
columns = ['tx_hash']
df.drop(columns, inplace=True, axis=1) #tx_hash column deleted because it is irrelevant for the case-scenario
```

```python
df
```

```python
proj_counts=df.nft_project_name.value_counts().reset_index()#number of nft projects transacted in the given time period
proj_counts.columns=['nft_project_name','Totalcount']
proj_counts
```

```python
proj_avg_amount=df.groupby('nft_project_name')['usd_amount'].apply(np.mean)    #Average amount for respective nft project transacted
proj_avg_amount.columns=['nft_project_name','avg_amount']
proj_avg_amount
```

```python
df3 = pd.merge(proj_counts, proj_avg_amount, on=['nft_project_name'])
print(df3)
df3.dtypes
```

```python
df3['trade_volume']=df3.Totalcount * df3.usd_amount

df3.sort_values(by=['trade_volume'],ascending=False) #NFT projects with highest total trade volume over the time period (total count*usd_amount)
```

```python
df2=df.groupby(['nft_project_name'])['usd_amount'].sum().sort_values(ascending=False).head(5) 
df2         
```
